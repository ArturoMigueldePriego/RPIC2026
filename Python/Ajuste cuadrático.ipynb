```python
import numpy as np
import matplotlib.pyplot as plt

# Datos: 6 pares (x, y)
x = np.array([60712.0, 101260.0, 144244.0, 180316.0, 219888.0])/1e6
y = np.array([3.0, 6.0, 10.0, 14.0, 19.0])*0.395/20

# Ajuste cuadrático (grado 2)
coef = np.polyfit(x, y, 2)

a, b, c = coef
print(f"y ≈ {a:.4f}x² + {b:.4f}x + {c:.4f}")

# Generar curva suave
x_fit = np.linspace(min(x), max(x), 100)
y_fit = a*x_fit**2 + b*x_fit + c

# Gráfica
plt.scatter(x, y, label="Datos")
plt.plot(x_fit, y_fit, label="Ajuste cuadrático")
plt.legend()
plt.grid()
plt.show()
```
